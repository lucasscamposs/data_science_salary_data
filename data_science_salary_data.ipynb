# Projeto de Ciência de Dados - Experiência e Salário dos Funcionários

### Tempo de Experiência vs Salário

O dataset que vamos explorar trata-se de dados fictícios do tempo de experiência dos funcionários e do salário que eles recebem.

O objetivo da análise consiste em responder: "Se uma empresa fosse contratar esses funcionários, o tempo de experiência que eles possuem fariam com que a empresa tivesse que pagar um salário maior por esses funcionários?"

Link dos dados: https://www.kaggle.com/datasets/karthickveerakumar/salary-data-simple-linear-regression/data



```python
# Bibliotecas

# Modelagem e Matrizez
import numpy as np
import pandas as pd

# Ler a Base de dados
import os

# Análises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
# Lendo a base de dados
os.chdir(r'C:\Users\Usuário\Desktop\Salary data - Simple linear regression')

Base_Dados = pd.read_csv('Salary_Data.csv')

# Verificando
Base_Dados.head()
```

```python
# Renomenado as colunas
Base_Dados.rename( columns={
    'YearsExperience' : 'Xp',
    'Salary' : 'Renda'
}, inplace=True )
```

```python
# dimensao
Base_Dados.shape
```

```python
# Campos nulos
Base_Dados.isnull().sum()
```

```python
# Grafico Nulos
sns.heatmap( Base_Dados.isnull(), cbar=False );
```

```python
Base_Dados.describe()
```

```python
plt.figure( figsize=(10, 6) )
plt.title('Análise da Renda')
sns.kdeplot( Base_Dados['Renda'], color='green' );
```

```python
plt.figure( figsize=(10, 6) )
plt.title('Análise da Xp')
sns.kdeplot( Base_Dados['Xp'], color='blue' );
```

```python
# Distribuição
plt.figure( figsize=(3, 5) )
sns.boxplot( Base_Dados['Renda'] );
```

```python
# Distribuição
plt.figure( figsize=(3, 5) )
sns.boxplot( Base_Dados['Xp'] );
```

```python
# Distribuição
plt.figure( figsize=(10, 5) )
sns.scatterplot( data=Base_Dados, x='Renda', y='Xp');
```

```python
plt.figure( figsize=(10, 5) )
sns.regplot( data=Base_Dados, x='Renda', y='Xp');
```

```python
# Correlação
Correlacao = Base_Dados.corr('pearson')
plt.figure( figsize=(10, 5) )
sns.heatmap( Correlacao, annot=True, cbar=True );
```

```python
# Correlação
Correlacao = Base_Dados.corr('spearman')
plt.figure( figsize=(10, 5) )
sns.heatmap( Correlacao, annot=True, cbar=True );
```
